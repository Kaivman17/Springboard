```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
engagement_df = pd.read_csv('takehome_user_engagement.csv')
users_df = pd.read_csv('takehome_users.csv', encoding='latin1')
```

```python
engagement_df.head()
```

```python
users_df.head()
```

```python
engagement_counts = {}

for index,row in engagement_df.iterrows():
    if row['user_id'] in engagement_counts:
        engagement_counts[row['user_id']] += 1
    else:
        engagement_counts[row['user_id']] = 1
```

```python
engagement_counts_df = pd.DataFrame.from_dict(engagement_counts, orient='index')
engagement_counts_df.columns= ['visit_count']
```

```python
engagement_counts_df = engagement_counts_df.reset_index(drop=False)
```

```python
engagement_counts_df.columns=['user_id', 'visit_count']
engagement_counts_df.head()
```

```python
adopted_user_engagement_counts = engagement_counts_df[engagement_counts_df['visit_count'] >= 3]
adopted_user_engagement_counts.head()
```

```python
adopted_user_engagement_df = engagement_df[engagement_df['user_id'].isin(adopted_user_engagement_counts['user_id'])]
```

```python
users_df['last_session_creation_time'] = pd.to_datetime(users_df['last_session_creation_time'], unit='s')
```

```python
users_df['creation_time'] = pd.to_datetime(users_df.creation_time)
```

```python
users_df['creation_time_difference'] = users_df['last_session_creation_time'].sub(users_df['creation_time'], axis=0)
users_df['creation_time_difference'] = users_df['creation_time_difference'] / np.timedelta64(1, 'D')
users_df.head()
```

```python
users_df = users_df[users_df.creation_time_difference > 7]
users_df.rename(columns={'object_id': 'user_id'}, inplace=True)
users_df.head(20)
```

```python
final_users_df = users_df[users_df.user_id.isin(adopted_user_engagement_counts.user_id)]
final_users_df.head(20)
```

```python
final_users_df.creation_source.value_counts()
```

```python
final_users_df.groupby('creation_source').size() * 100 / len(final_users_df)
```

For adopted users, the most common form of account creation appears to be being invited by an organization. About 35% of adopted users were invited by an organization. 22% of adopted users were invited by a guest, 18% of adopted users signed up via the website, 14% signed up using google authentication, and 10% are doing personal projects.

```python
final_users_df.opted_in_to_mailing_list.value_counts()
```

Interestingly, it appears that most adopted users have not opted into receiving marketing emails.

```python
final_users_df.enabled_for_marketing_drip.value_counts()
```

It also appears that most adopted users are not on the regular marketing email drip

# Conclusion

It appears that most adopted users are developed through references less so through the company's marketing strategy. To continue increasing the amount of adopted users moving forward I would recommend that the company provides incentives for current adopted users to refer other potential users and also alter the email marketing strategy as it has not shown to be effective thus far.
